Converting to Database

```python
import sqlite3

conn = sqlite3.connect("../database/food_wastage.db")

providers.to_sql(
    "providers",
    conn,
    if_exists="replace",
    index=False
)

receivers.to_sql(
    "receivers",
    conn,
    if_exists="replace",
    index=False
)

food.to_sql(
    "food_listings",
    conn,
    if_exists="replace",
    index=False
)

claims.to_sql(
    "claims",
    conn,
    if_exists="replace",
    index=False
)

conn.close()
```

Provider Type Distribution

```python
provider_type = providers["Type"].value_counts()

print(provider_type)

provider_type.plot(
    kind="bar"
)

plt.title("Provider Type Distribution")
plt.xlabel("Provider Type")
plt.ylabel("Number of Providers")
plt.xticks(rotation=0)

plt.show()
```

Food Quantity Distribution

```python
plt.figure(figsize=(10,5))

plt.hist(
    food["Quantity"],
    bins=8,
    rwidth=0.9       # small gap between bars
)

plt.title("Food Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

Top 10 Providers By Total Donation

```python
top_providers = (
    food.groupby("Provider_ID")["Quantity"]
    .sum()
    .reset_index()
)

top_providers = top_providers.merge(
    providers[["Provider_ID","Name"]],
    on="Provider_ID"
)

top10 = top_providers.sort_values(
    by="Quantity",
    ascending=False
).head(10)

print(top10)
```

```python
plt.figure(figsize=(12,5))

plt.bar(
    top10["Name"],
    top10["Quantity"]
)

plt.title("Top 10 Providers By Total Donation")
plt.xlabel("Provider")
plt.ylabel("Quantity Donated")

plt.xticks(rotation=60)

plt.show()
```

Top 10 Most Claimed Food Items

```python
top_foods = (
    claims.groupby("Food_ID")
    .size()
    .reset_index(name="Claims")
)

top_foods = top_foods.merge(
    food[["Food_ID","Food_Name"]],
    on="Food_ID"
)

top_foods = top_foods.sort_values(
    by="Claims",
    ascending=False
).head(10)

print(top_foods)
```

```python
plt.figure(figsize=(12,5))

plt.bar(
    top_foods["Food_Name"],
    top_foods["Claims"]
)

plt.title("Top 10 Most Claimed Food Items")
plt.xlabel("Food Item")
plt.ylabel("Number of Claims")

plt.xticks(rotation=60)

plt.show()
```
